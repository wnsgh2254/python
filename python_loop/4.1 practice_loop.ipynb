# 문제 1.(for)
## for 문을 이용하여 1부터 입력받은 정수까지의 합을 구하여 출력하는 프로그램을 작성하시오.
- 100 이하의 양의 정수만 입력된다. 
- 입력 예: 10
- 출력 예: 55

```python
num = int(input("양의 정수를 입력하세요 : ")) # 
add = 0    # for문을 활용하기 위해 add 변수를 int 형태로 지정함. 
for i in range(num+1): # range(10) 는 1~9 까지 이기 때문에 num+1 을 함으로써 원하는 결과를 얻을 수 있다
  add = add+i # for문을 통해 가져온 숫자를 add 변수에 계속 더해줌

print(add) # 반복문 밖에서 출력하는 이유는 출력 예제에 맞추기 위함
```

# 문제 2.
## 정수를 계속 입력받다가 100 이상의 수가 입력이 되면 마지막 입력된 수를 포함하여 합계와 평균을 출력하는 프로그램을 작성하시오.

- 평균은 반올림하여 소수 첫째자리까지 출력한다.

```python
num_list = [] # 받은 숫자를 저장해둘 리스트 
add = 0   # 반복문 안에서 만들어도 되지만 반복문을 간단하게 만들기 위해 밖에다 지정

while True: # 반복문을 끝내도록 명령(False or break)하지 않으면 계속 진행
  num = int(input("정수를 입력하세요 : "))
  num_list.append(num)  # num 이 입력받은 숫자를 num_list에 저장하는 함수를 사용
  if num >= 100:
    for i in num_list:
      add = add+i # 수업시간에 배움
    break
avg = add/ len(num_list) # 평균을 구하는 변수
print("합계 : {} ".format(add))
print("평균 : {} ".format(round(avg, 1)))   # round는 반올림을 해주며 몇번 째 자리까지 출력할지 지정할 수 있음
```

# 문제 3.
## 정수를 입력받아서 3의 배수가 아닌 경우에는 아무 작업도 하지 않고, 3의 배수인 경우에는 3으로 나눈몫을 출력하는 작업을 반복하다가 -1이 입력되면 종료하는 프로그램을 작성하시오.
예시:
- 5
- 12 
- <h2>4</h2>
- 21
- <h2>7</h2>
- 100
- -1

```python
while True:
  num = int(input("정수를 입력하세요 : "))
  if num == -1:
    break
  elif num % 3 != 0:
    continue
  elif num % 3 == 0:
    print(num//3)
    continue
```

# 보너스(for)
## 아래와 같은 모양으로 출력하는 프로그램을 작성하시오.
 - 사용자에게 숫자를 입력받아 입력 받은 높이 만큼의 삼각형을 출력하시오.
 - 입력 예: 5

```python
    *
   **
  ***
 ****
*****
```

```python
num = int(input("숫자를 입력하세요 : "))

for i in range(num+1): #  갯수를 늘려나가며 모양을 맞추기 위해 for문을 사용
  star = i*"*" # i 가 커질수록 * 가 많아짐
  print(star.rjust(num)) # for문을 이용해 star을 출력하는데 오른쪽 정렬이며 확보할 공간을 입력받은 num 만큼으로 지정
```

```python
country = {1:"서울(Seoul)", 2:"워싱턴(Washington)", 3:"동경(Tokoy)",4:"북경(Beijing)"} # 딕셔너리를 이용해 key 를 입력하면 value를 출력할 수 있다.


while True:
    num = int(input("1. Korea\n2. USA\n3. Japan\n4. China number? : ")) # 예시와 맞추기 위해 줄바꿈으로 \n을 사용
    if num > len(country):
        print("none")
        break
    print(country[num]) # country 딕셔너리의 num 이 키값으로 되어있는 벨류를 출력
    continue
```

# 문제 5.
## 0 이상의 정수들이 공백으로 구분되어 반복적으로 주어진다. 0이 입력되면 반복문을 멈추고 그 전까지 입력받은 수들에 대하여 홀수의 개수와 짝수의 개수를 출력하는 프로그램을 작성하시오.
- 입력 예: 9 7 10 5 33 65 0
- 출력 예: 
 odd : 5
 even : 1

```python
odd = []
even = []
while True:
    num = int(input("정수를 입력하세요 : "))
    num_list = [] 
    num_list.append(num) # num 에서 받은 숫자를 num_list에 저장
    if num == 0: # 0이 나오면 반복문 종료 후 맨 밑에있는 print문을 출력
        break
    for i in num_list:
        if i % 2 != 0: # i 를 2로 나눈 값의 나머지가 0 이 아니라면 == 짝수가 아니라면
            odd.append(i)     # 위에 지정해둔 odd 변수에 넣는다                   
        elif i % 2 == 0: # i 를 2로 나눈 값의 나머지가 0 이면 변수 even에 추가한다 (짝수를 정의하는 것을 표현하기 위해 else 말고 문장을 썼음.)
            even.append(i)
           


print("odd : {} even : {}".format(len(odd),len(even))) # list 형태로 저장됐기 때문에 len를 통해 몇개가 저장됐는지 알 수 있음.
```

# 문제 6.
## 삼각형의 밑변의 길이와 높이를 입력받아 넓이를 출력하고, "Continue? "에서 하나의 문자를 입력받아 그 문자가 'Y' 나 'y' 이면 작업을 반복하고 다른 문자이면 종료하는 프로그램을 작성하시오.

- 넓이는 반올림하여 소수 첫째자리까지 출력한다.
예시
###### Base = 11
###### Height = 5
###### Triangle width = 27.5
###### Continue? Y
###### Base = 10
###### Height = 10
###### Triangle width = 50.0
###### Continue? N

```python
while True:
  base = int(input("밑변의 길이를 입력하세요 : "))
  height = int(input("높이를 입력하세요 : "))
  width = round((base*height)/2, 1)
  print("삼각형의 넓이 {}".format(width))
  work = input("Continue? : ")
  if work == "y" or work == "Y":
    continue
  else:
    break
```

# 문제 7.(for)
## 100 이하의 두 개의 정수를 입력받아 작은 수부터 큰 수까지 차례대로 출력하는 프로그램을 작성하시오.
- 입력 예: 5 10
- 출력 예: 5 6 7 8 9 10

```python
sort = []
while len(sort) <= 2: # 입력받은 정수의 갯수가 2개를 넘어가기 전 까지 밑에 실행문들을 실행
  num1 = int(input("100 이하의 숫자를 입력하세요 : "))
  num2 = int(input("100 이하의 숫자를 입력하세요 : "))

  if num1 > 100 or num2 > 100:
    print("100이하의 숫자만 입력하세요")
    continue
  elif num1 > num2: 
    for i in range(num2,(num1+1)):   # range(1,10) 은 1~9 까지 이기 때문에 뒤에 올 숫자에 +1 해주었다
      sort.append(i)
  else:
    for i in range(num1, (num2+1)): # num1 과 num2 중 무엇이 더 큰 숫자일지 모르기 때문에 else를 사용
      sort.append(i)
  break

print(sort)
```

# 문제 8.(for)
## 두 개의 정수를 입력받아 두 정수 사이(두 정수를 포함)에 3의 배수이거나 5의 배수인 수들의 합과 평균을 출력하는 프로그램을 작성하시오.
- 평균은 반올림하여 소수 첫째자리까지 출력한다.
- 입력 예: 10 15
- 출력 예: 
sum : 37
avg : 12.3

```python
num1 = int(input("정수를 입력하세요 : "))
num2 = int(input("정수를 입력하세요 : "))
num_list = [] # 평균을 구할때 갯수를 알아야 하기 때문에 리스트를 만들었음
sum = 0
if num1 > num2:
  for i in range(num2,(num1+1)):
    if i % 3== 0 or i % 5 == 0: # 3 또는 5의 배수라면
      num_list.append(i) # 리스트에 넣는다
      sum = sum + i
else: # num1 과 num2 중 무엇이 클지 모르기 때문에 작성
  for i in range(num1,(num2+1)):
    if i % 3 == 0 or i % 5 == 0:
      num_list.append(i)
      sum = sum + i
avg = round(sum/len(num_list), 1) # round 로 1번째 자리까지 반올림

print("sum : {} avg : {}".format(sum,avg))
```

# 문제 9. (for)
## 2부터 9까지의 수 중 2개를 입력받아 입력받은 수 사이의 구구단을 출력하는 프로그램을 작성하시오.
- 단 반드시 먼저 입력된 수의 구구단부터 아래의 형식에 맞게 출력하여야 한다.
- 구구단 사이의 공백은 3칸이다.
- 입력 예: 5 3
- 출력 예:
5 * 1 =  5   4 * 1 =  4   3 * 1 =  3
5 * 2 = 10   4 * 2 =  8   3 * 2 =  6
5 * 3 = 15   4 * 3 = 12   3 * 3 =  9
5 * 4 = 20   4 * 4 = 16   3 * 4 = 12
5 * 5 = 25   4 * 5 = 20   3 * 5 = 15
5 * 6 = 30   4 * 6 = 24   3 * 6 = 18
5 * 7 = 35   4 * 7 = 28   3 * 7 = 21
5 * 8 = 40   4 * 8 = 32   3 * 8 = 24
5 * 9 = 45   4 * 9 = 36   3 * 9 = 27

```python
while True:
  num1 = int(input("정수를 입력하세요 (2~9): "))
  num2 = int(input("정수를 입력하세요 (2~9): "))
  if (num1 < 2 or num1 > 9) or (num2 < 2 or num2 > 9): # 2~9 사이의 숫자 조건을 만족 시키기 위한 문장
    print("2 와 9 사이의 숫자를 입력하세요")
    continue
  elif num1 > num2: # range를 사용할때 어떤 숫자가 더 큰지 알아야 하기 때문에 크기를 비교
    for i in range(num1,(num2-1),-1): # 몇단을 구구단 할지 정함
      for x in range(1,10): # 구구단중 1~9 역할을 해줌
        print("{}*{}={}".format(i,x,(i*x)), end="   ")
  
  else:
    for i in range(num1,(num2+1)):
      for x in range(1,10):
        print("{}*{}={}".format(i,x,(i*x)), end="   ")
  
  break
```

 - 입력 예: 5

```python
*****
 ****
  ***
   **
    *
```

```python
num = int(input("숫자를 입력하세요 : "))

for i in range(num,0,-1): #  반복 할때마다 -1씩 됨
  star = i*"*" 
  print(star.rjust(num)) # for문을 이용해 star을 출력하는데 오른쪽 정렬이며 확보할 공간을 입력받은 num 만큼으로 지정
```
