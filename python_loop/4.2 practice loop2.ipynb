# 문제 1.
## 1부터 100까지의 정수 중 한 개를 입력받아 100 보다 작은 배수들을 차례로 출력하다가 10 의 배수가 출력되면 프로그램을 종료하도록 프로그램을 작성하시오.
- 입력 예: 7
- 출력 예: 7 14 21 28 35 42 49 56 63 70

```python
num = int(input("정수를 입력하세요 : "))
num_val = 100//num
 
for i in range(1,num_val+1):
  x = i*num
  time = ""
  time += str(x)
  print(time, end=" ")
  if x > 100 or x % 10 == 0:
    break
```

# 문제 2.

```python
    1
  1 2
1 2 3

    1
  1 3
1 3 5
    
```

```python
n = int(input("숫자를 입력하세요 : "))

for x in range(1,n+1): # 3 을 넣을경우 1,2,3 의 range를 갖음
  a = 1
  for y in range(1,n-x+1): # n에 5 를 넣을 경우 x 문이 1바퀴 돌때 y는 range(1,4) 를 돈다. x가 커질수록 공백을 출력하는 수가 줄어듬
    print("  ",end='')
  for i in range(1,x+1): # 공백 뒤에 올 숫자를 뽑을 반복문
    print(a, end=' ') # 첫 번째 바퀴에 range는 1,2 이므로 1만 출력.
    a+=1 # i 반복문 안에서 a +=1 을 해줌으로써 1 12 123 이 출력 가능
  print() 

print()

for x in range(1,n+1):
  a = 1
  for y in range(1,n-x+1):
    print("  ",end='')
  for i in range(1,x+1):
    print(a,end=' ')
    a+=2  # 1,2,3 뽑는 코드와 같지만 a에 2씩 더해줘서 1 3 5 가 출력됨
  print()

```

# 문제 4.
## 10개의 정수를 입력받아 100 미만의 수 중 가장 큰 수와 100 이상의 수 중 가장 작은 수를 출력하는 프로그램을 작성하시오.
- List 데이터 구조를 활용하시오.
- 입력 예: 88 123 659 15 443 1 99 313 105 48
- 출력 예: 99 105

```python
num_list = []
num_sml = []
num_big = []

while len(num_list) < 10:
  num = int(input("정수를 입력하세요 : "))
  num_list.append(num) # 입력 받은 정수를 저장
  for i in num_list:
    if i < 100:          # 받은 정수가 100 보다 작다면 만들어 놓은 sml리스트에 저장
      num_sml.append(i)
    else:
      num_big.append(i)   #받은 정수가 100보다 크다면 big 리스트에 저장

print("100이상의 수중에 가장 작은 수 : {}".format(min(num_big))) # big 리스트 (100이 넘음) 중에 가장 작은 숫자를 뽑아내는 min 을 사용
print("100미만의 수중에 가장 큰 수 : {}".format(max(num_sml))) # sml 리스트 (100 미만) 중에 가장 큰 숫자를 뽑아내는 max를 사용
```

## 사용자에게 base문자열을 첫 줄에 입력받고,  다음으로 찾고자하는 문자 1개를 입력받아 문자열 내의 검색 문자의 모든 위치를 출력하는 프로그램을 작성하시오.
- 첫 번째 위치는 0번이며 리스트에 없는 문자가 입력되면 "none" 라는 메시지를 출력하고 끝내는 프로그램을 작성하시오.
- 예제: "my name is dean" 을 base 문자열로 입력하고, 찾고자 하는 문자를 'd'로 입력하면 11을 출력해주면 된다.
- 만약, 찾으려는 문자가 2개 이상인 경우에는 모든 위치를 출력해준다.

```python
name = input("문자열을 입력하세요 : ")
letter = input("찾고 싶은 문자 : ")
index_list = []
num = 0

for i in name:
  if i == letter:
    index_list.append(name.index(i,num))
  num += 1

print(index_list)
```

# 문제 7.
## 학생들의 점수를 입력을 받다가 0이 입력되면 그 때까지 입력받은 점수를 10점 단위로 구분하여 점수대별 학생 수를 출력하는 프로그램을 작성하시오. 
- 1명도 없는 점수는 출력하지 않는다.
- 입력 예: 63 80 95 100 46 64 88 0

```python
출력 예: 
100 : 1 person
90 : 1 person
80 : 2 person
60 : 2 person
40 : 1 person
```

```python
score_list = []

while True:
  score = int(input("점수를 입력하세요 : "))
  if score == 0:
    break
  score_list.append((int(score * 0.1))*10) # 리스트에 넣은 숫자들의 1의자리 숫자들을 없애기 위해 10으로 나눈 숫자를 int로 바꾼 뒤에 10을 다시 곱해주었다

set_score = sorted(set(score_list), reverse=True) # score_list를 내림차순으로 정렬함
person = [score_list.count(a) for a in set_score] # set_score 를 a라는 변수로 반복한뒤, a가 set_score에 들어있는 갯수를 person이라는 변수에 넣는다
d = dict(zip(set_score,person)) # zip함수로 숫서에 맞게 짝을 지어준 뒤에 딕셔너리 형태로 묶음.

for x,y in d.items(): # items 는 딕셔너리 형태인 d의 키와 벨류를 튜플 형식으로 묶어준다. items 을 쓰지 않을 경우 key 만 가져옴.
  print("{}:{} person".format(x,y))
```

# 문제 8.
## 소수(prime number)란 1보다 큰 자연수 중 1과 자기 자신 두 개만을 약수로 갖는 수를 말한다. 
- 자연수 M과 N을 입력받아 M부터 N까지 소수의 개수를 구하여 출력하는 프로그램을 작성하시오.
- M이상 N이하의 자연수 중 소수가 몇 개인지 구하여 출력한다.
- 입력 예: 10 100
- 출력 예: 21

```python
m = int(input("자연수를 입력하세요 : "))
n = int(input("자연수를 입력하세요 : "))
print_num = []

for i in range(m,n+1):
  if i % 2 != 0 and i % 3 != 0 and i % 5 != 0 and i % 7 != 0: # 좋은 방법은 아닌 것 같지만 생각이 안나서.. 2,3,5,7로 나누어지지 않으면 소수이기 때문에 이런 방법을 썼다..
    print_num.append(i)
print(len(print_num)) # 리스트에 소수들이 들어가서 그 리스트의 길이가 갯수가 된다
```

# 보너스
## 자연수 M과 N이 주어질 때 M이상 N이하의 자연수 중 소수인 것을 모두 골라 이들 소수의 합과 최소값을 찾는 프로그램을 작성하시오.

- 예를 들어 M=60, N=100이 경우 60이상 100이하의 자연수 중 소수는 61, 67, 71, 73, 79, 83, 89, 97 총 8개가 있으므로, 
- 이들 소수의 합은 620이고, 최소값은 61이 된다.

```python
m = int(input("자연수를 입력하세요 : "))
n = int(input("자연수를 입력하세요 : "))
print_num = []
add_num = 0
for i in range(m,n+1):
  if i % 2 != 0 and i % 3 != 0 and i % 5 != 0 and i. % 7 != 0:
    print_num.append(i)
    add_num += i
print("소수의 합 : {}".format(add_num))
print("최소값 : {}".format(min(print_num)))
```

# 문제 11.

- 입력 예: 10101
- 출력 예: 21

```python
n = input("2진수를 입력하세요 : ") # 2진수
num = len(n)-1 # 자릿값
a = 0
result = 0
for i in n:
  if i =='1':
    result += 2**(num-a)
  a+=1

print(result)
```

# 문제 12.

## 복호화 키와 암호 문자를 입력으로 받아 원문을 구하는 프로그램을 구현하시오. 복호화 키는 26개의 소문자로 주어지며, a,b,c,d... 를 순서대로 복호화 키 문자로 대치한다는 것을 뜻한다.

 

- 예를 들어, 복호화 키가 "eydbkmiqugjxlvtzpnwohracsf" 와 같이 주어진다고 하자, 

- 그러면 이는 다음과 같다 - a 문자는 e, b 문자는 y, ..., z 문자는 f로 바꿔줌.

- 암호화 된 문자는 대소문자 혹은 공백이 올수 있고 대소문자는 대문자로 소문자는 소문자로 치환 규칙에 맞게 출력하고 , 

- 공백문자는 그대로 출력한다.

- 입력 예: 
- eydbkmiqugjxlvtzpnwohracsf
- Kifq oua zarxa suar bti yaagrj fa xtfgrj
- 출력 예:
- Jump the fence when you seeing me coming

```python
string = input("문장을 입력하세요 : ")
key = input("복호화 문자를 입력하세요 :")
alp = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"

key2 = key.upper()
key3 = key + key2
x = string.maketrans(alp,key3)
print(string.translate(x))
```

# diamond code

```python
n = int(input("min : "))   # 입력한 수가 가운데 줄이 됨
a = 1
for i in range(n,-1,-1):
  print(i*" ", end='')
  print("*"*a)
  a += 2
a = a-4   # 위에서 가장 긴 줄을 출력하고 마지막에 +2 된 상태이기 때문에 이어가기 위해 -4를 해줌

for i in range(1,n+1):
  print(" "*i,end='')
  print("*"*a)
  a-=2
```
