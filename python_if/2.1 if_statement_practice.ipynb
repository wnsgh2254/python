# 문제 2.
- “몸무게+100-키”를 비만수치 공식이라고 하자.
- 키와 몸무게를 자연수로 입력받아 첫 번째 줄에 비만수치를 출력하고, 비만수치가 0보다 크면 다음줄에 비만("Obesity")이라는 메시지를 출력하는 프로그램을 작성하시오.
- 입력 예: 155 60
- 출력 예: 
    * 5
    * Obesity

```python
weight = int(input("몸무게를 입력하세요 : "))
height = int(input("키를 입력하세요 : "))
obesity = weight+100-height # 입력받은 두 변수의 합을 obesity 로 지정한다
if obesity > 0:   # obesityrk 0보다 크다면 
  print(obesity)  # obesity 값을 출력하고
  print("Obesity")  # obesity라는 문자를 출력한다
else:
  print(obesity)
```

# 문제 4.
- 복싱체급은 몸무게가 50.80kg 이하를 Flyweight 61.23kg 이하를 Lightweight, 72.57kg 이하를 Middleweight, 88.45kg 이하를 Cruiserweight, 88.45kg 초과를 Heavyweight라고 하자.

- 몸무게를 입력받아 체급을 출력하는 프로그램을 작성하시오.
- 입력 예: 87.3
- 출력 예: Cruiserweight

```python
weight = float(input("몸무게를 입력하세요 : "))
if weight <= 50.80:    # 입력받은 몸무게가 50.80보다 작거나 같다면
  print("Flyweight")   # Flyweight를 출력한다
elif weight <= 61.23:   # 위 조건에 만족하지 못한다면 이 조건으로 다시 물어본다
  print("Lightweight")  # 조건에 만족한다면 Lightweight 를 출력한다
elif weight <= 72.57:
  print("Middleweight")
elif weight <= 88.45:
  print("Cruiserweight")
else:
  print("Heavyweight")
```

# 문제 8.
- 년도를 입력받아 윤년(leap year)인지 평년(common year)인지 판단하는 프로그램을 작성하시오.
- 400으로 나누어떨어지면 윤년이다. 또는 4로 나누어떨어지고 100으로 나누어떨어지지 않으면 윤년이다. 나머지는 모두 평년이다.
- 입력 예: 2008
- 출력 예: leap year

```python
year = int(input("년도를 입력하세요 : "))
if year % 400 == 0:   # year 를 400으로 나눠서 나머지가 0 이라면 
  print("leap year")  #leap year 을 출력한다
elif year % 4 == 0 and year % 100 != 0: # year 가 4로 딱 떨어지고 동시에 year가 100으로 딱 나눠지지 않는다면,
  print("leap year") # leap year을 출력한다
else:
  print("common year")  #나머지의 경우 common year 을 출력한다
```

# 문제 10.
- 공백을 포함한 문자열을 입력받아 각 단어로 분리하여 문자열 배열에 저장한 후 입력순서의 반대 순서로 출력하는 프로그램을 작성하시오. 문자열의 길이는 100자 이하이다.
- 입력 예: C++ Programing jjang!!
- 출력 예:
    * jjang!!
    * Programing
    * C++

```python
while True: # 조건중에 문자열의 길이는 100자 이하이다 를 만족 시키기 위해 while문으로 시작 했습니다.
  word = input("공백을 포함한 문자열을 입력하세요 : ").split(" ") #word 라는 변수에 문자열을 입력하고 split 함수에 공백을 넣어, 공백 단위로 끊어서 리스트 형태로 입력했습니다. (순서를 거꾸로 바꾸기 위함)
  if len(word) <= 100: # 입력한 문자가 100자 이하라면, 반복문에서 빠져나갈 수 있도록 했습니다.
    break # 반복문 빠져나감
  else:
    print("문자열을 100자 이하로 입력해주세요.")
    continue # 문자열의 길이가 100이 넘었을경우 다시 input을 받도록 continue를 사용했습니다.

word.reverse() #split으로 인하여 word는 리스트 형태여서 reverse() 를 사용하여 리스트의 순서를 거꾸로 바꿨습니다.
for i in word: # 세로 형태로 하나 하나 출력하기 위해 for문을 이용했습니다,
  print(i)
```
