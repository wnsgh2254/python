# 문제 1.
- "FizzBuzz"문제 
 - 1에서 100까지의 숫자를 출력하되, 
 - 만약 해당 숫자가 '3'으로 나누어지면 숫자 대신에 "Fizz"를 출력하고, 
 - 만약 해당 숫자가 '5'로 나누어지면 숫자 대신에 "Buzz"를 출력하며, 
 - 만약 해당 숫자가 '3'과 '5'로 모두 나누어지면 숫자 대신에 "FizzBuzz"를 출력
 - 입력 예1: 5
 - 출력 예1: "Buzz"
 - 입력 예2: 3
 - 출력 예2: "Fizz"
 - 입력 예3: 15
 - 출력 예3: "FizzBuzz" 

```python
while True: # 100이하의 숫자 조건을 만족하기 위해 while문으로 시작했다
  num = int(input("숫자를 입력하세요(100이하) : "))
  if num <= 100:
    break # 조건에 만족할경우 반복문을 빠져나가 나머지 조건을 충족시킨다
  else:
    print("숫자를 100 이하로 입력해주세요")
    continue

if num % 3 == 0 and num % 5 == 0 : # 출력 조건중 3번째 이지만, 파이썬은 위에서 아래로 진행되기 때문에 이 자리에 Fizz 조건을 넣는다면, 15을 입력해도 FizzBuzz가 아닌 Fizz만 출력되기 때문에 순서를 바꿈
  print("FizzBuzz")
elif num % 5 == 0:
  print("Buzz")
elif num % 3 == 0:
  print("Fizz")
else:
  print(num) # 아무 조건에 만족되지 못한다면, 그냥 입력받은 숫자를 출력
```

# 문제 2.
- 나이를 입력받아 나이에 따라 지하철 교통요금 출력하는 프로그램을 만드시오.
- 8세 미만의 어린이의 지하철 교통요금은 450원이다.
- 8세 이상 19세 이하의 청소년 지하철 교통요금은 720원이다.
- 20세 이상의 성인 지하철 교통요금은 1250원이다
- 입력 예1: 2
- 출력 예1: "450원입니다."
- 입력 예2: 19
- 출력 예2: "720원입니다."
- 입력 예3: 82
- 출력 예3: "1250원입니다."

```python
age = int(input("나이를 입력하세요 : "))

if age < 8:
  print("450원 입니다.")
elif 8 <= age <= 19:
  print("720원 입니다.")
else:
  print("1250원 입니다.")
```

# 문제3.
- 초를 입력 받아서 시간과 분으로 변환하는 프로그램을 작성하시오.
- 입력 예: 3598
- 출력 예: 59분 58초

```python
num = int(input("초를 입력하세요 : "))
min = num//60 # 입력받은 초를 60으로 나눈 몫 == 분
sec = num%60 # 입력받은 초를 60으로 나눈 나머지 == 초

print("{}분 {}초".format(min, sec))
```

# 문제 4.

- 수식:  F(화씨) = 9.0 / 5.0 * C(섭씨)+ 32

# 문제 7.
- 가상의 두 사람(A, B)이 가위 바위 보 중 하나를 내서 승패를 가르는 가위 바위 보 게임을 작성하십시오.
- 이 때 ["가위", "바위", "보"] 리스트를 활용합니다.
- 입력 예: "가위" "바위"
- 출력 예: "B가 이겼습니다."

```python
while True: # for문을 이용하면 간단하게 끝나지만 일어날만한 상황들을 위해 while을 사용했습니다.
  game = ["가위","바위","보"]
  a = input("A님 무엇을 내시겠습니까? : ")
  b = input("B님 무엇을 내시겠습니까? : ")

  if not (a or b) in game: # 
    print('"가위","바위","보" 중에서 입력해주세요')
    continue
  elif a == b:
    print("비겼습니다")
  elif a == game[0] and b == game[1]:   # 조건중에 리스트를 활용하라는 것이 있어서 변수에 넣어 index를 이용해 풀었습니다. 
    print("B가 이겼습니다.") 
  elif a == game[1] and b == game[2]:
    print("B가 이겼습니다.")
  elif a == game[2] and b == game[0]:
    print("B가 이겼습니다,")
  else:
    print("A가 이겼습니다.")
  
  break
```

# 문제 8.
- 다음의 결과와 같이 입력된 문자가 대문자일 경우 소문자로, 소문자일 경우 대문자로 변경하고,
- 알파벳이 아닐 경우엔 그냥 출력하는 코드를 작성하십시오.
- 입력 예1: A
- 출력 예1: a
- 입력 예2: 가나다
- 출력 예2: 가나다


```python
string = input("문자를 입력하세요 : ")

if string.isalpha() == True and string == string.upper():
  print(string.lower())
elif string.isalpha() == True and string == string.lower():
  print(string.upper())
else:
  print(string)
```
